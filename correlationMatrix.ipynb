**Correlation Matrix**

```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('df_train.csv')
df
```

```python
corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
sn.set(rc={'figure.figsize':(25,15)})
plt.show()
```
