## Visualizing Missing Values

```python
import pandas as pd
import missingno as msno
```

```python
# !pip install missingno
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("train.csv")
```

MATRIX: Using this matrix you can very quickly find the pattern of missingness in the dataset. 

```python
msno.matrix(df_train)
```

```python
msno.matrix(df_test)
```

BAR CHART: This bar chart gives you an idea about how many missing values are there in each column

```python
msno.bar(df_train)
```

```python
msno.bar(df_test)
```

HEATMAP: Heatmap shows the correlation of missingness between every 2 columns.

A value near -1 means if one variable appears then the other variable is very likely to be missing.
A value near 0 means there is no dependence between the occurrence of missing values of two variables.
A value near 1 means if one variable appears then the other variable is very likely to be present.

```python
msno.heatmap(df_train)
```

```python
msno.heatmap(df_test)
```

DENDROGRAM: The dendrogram allows you to more fully correlate variable completion, 
revealing trends deeper than the pairwise ones visible in the correlation heatmap

```python
msno.dendrogram(df_train)
```

```python
msno.dendrogram(df_test)
```
